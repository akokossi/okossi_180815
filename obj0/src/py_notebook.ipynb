```
import sys
from pyspark import * 
from pyspark.sql.functions import *
import urllib2 
from operator import add
from pyspark.sql import SparkSession

target_url="https://raw.githubusercontent.com/ToJen/Quorum-Enterprise-Blockchain/master/README.md"
data = urllib2.urlopen(target_url).read()
for char in '#.,\n':
    data=data.replace(char,' ')
data = data.lower()
data = data.split()
lines = sc.parallelize(data)  
counts = lines.flatMap(lambda x: x.split(' ')) \
                  .map(lambda x: (x, 1)) \
                  .reduceByKey(add)
output = counts.collect()  
nWords=0
print "WORD                           COUNT"
print "===================================="
for (word, count) in output:
  nWords=nWords+1   
  print("%s                                %i" % (word, count))
  
print "====================================="
print
print
print "Total words count=" + str(nWords)
```
